```python
#!/usr/bin/env python3
#-*- coding:utf-8 -*-
'''
利用高德地图api查询地点
'''
import requests
import pprint

def search_place(p_keyword, p_city, p_page ):
    parameters = {
        'keywords': p_keyword, 
        'city': p_city , 
        'output': "json",
        'offset': 25 , 
        'page': p_page , 
        'citylimit':"true",
        'key': '58488a2b3762ba1ac6f5f4f2a176c553'}
    base = 'https://restapi.amap.com/v3/place/text'
    response = requests.get(base, parameters)
    answer = response.json()
    return answer["pois"]
```

```python
# adcode,城区
cq_district = ( 
    500105 ,#,"江北区"
    500103 ,#,"渝中区"
    500104 ,# ,"大渡口区"
    500107 ,#,"九龙坡区"
    500112 ,#,"渝北区"
    500113 ,#,"巴南区"
    500108 ,#,"南岸区"
    500106 ,#,"沙坪坝区"
    500109 )#,"北碚区"
```

```python
search_place("火锅",cq_district[1],10)
```

```python
all_lists = []

for adcode in cq_district:
    flag_w = 1
    page_w = 1
    while flag_w :
        # search & append
        print(adcode,page_w)
        add_list = search_place("火锅",adcode,page_w)
        all_lists = all_lists + add_list       
        page_w += 1
        
        if len(add_list) == 0:
            flag_w = 0
            page_w = 1
        
```
